```python
#TRUE CASE
# Example_1 [Condition satisfy true]
name = "python"       #[= means assigining]
if name == "python":  #[== means condition]
    print('won')
```

```python
#FALSE CASE
#Example_2 [Condition not satisfying]
name = "python"
if name == "py":    #[Condition is false so it is not printing] 
    a=12
    b=21
    print(a+b)
```

```python
#Example_3  [if and else in one code]
name = "python"
if name == "python":        #[Condition is true ]
    print("won")
else:
    a = 12
    b = 32
    print(a+b)
```

```python
#Example_4  [if and else in one code]
name = "python"
if name == "py":        #[Condition is false ]
    print("won")
else:
    a = 12
    b = 32
    print(a+b)        #[Condition is false, so else is executed]
```

```python
#Example_3 [Condition by using user input] [STEP BY STEP ANALYSIS ALSO]
print("Hello")
print(10)
num_1 = eval(input("Enter the first number to be calculate"))
num_2 = eval(input("Enter the second number to be calculate"))
if num_1==num_2:
    print("Condition is True case")
    print("We are doing addition operartion")
    print("The addition of {} and {} is {} ".format(num_1,num_2,num_1+num_2))
else:
    print("Condition is False case")
    print("We are doing multiplication operation")
    print("The multiplication of {} and {} is {}".format(num_1,num_2,num_1*num_2))
substraction = num_1 - num_2
print("Subraction",substraction)
print("Great!")
```
