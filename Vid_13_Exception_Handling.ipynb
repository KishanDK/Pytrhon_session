```python
#Exception Handling : There are two blocks in exception handling[try: and except:]
#try: - All the code is in try block 
#except - The error line is in except block
```

```python
#Example_1
try:
    num_1 = eval(input("Enter the number 1:"))
    num_2 = eval(input("Enter the number 2:"))
    Sum = num_1 + num_2
    print(Sum)
except:
    print("There is an error, check the code once")      #[There is no error so it will give output]
```

```python
# Example_6 [Square of a number]
try:
    num_1 = eval(input("Enter the number 1:"))
    Squared_num = num_1 * num_1
    print("The square of number {} is {} ".format(num_1,Squared_num))
except Exception as e:
    print(e)
```
