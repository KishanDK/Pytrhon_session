```python
import math
from collections import deque

def shortest_referral_chain(a, b, c):
    n = a + b + c
    if 2 * a + b != n - 1:
        return -1  # Not a valid tree

    # Start BFS simulation from root
    queue = deque()
    queue.append(1)  # Start with root (1 active referrer)
    level = 0

    # Intern pool (excluding root)
    total_needed = n - 1

    while total_needed > 0:
        level += 1
        next_level = 0
        while queue and (a > 0 or b > 0) and total_needed > 0:
            curr = queue.popleft()
            for _ in range(curr):
                if a > 0:
                    queue.append(2)
                    next_level += 2
                    a -= 1
                    total_needed -= 2
                elif b > 0:
                    queue.append(1)
                    next_level += 1
                    b -= 1
                    total_needed -= 1
                else:
                    break
        if next_level == 0 and total_needed > 0:
            return -1
    return level

# Read input
T = int(input())
for _ in range(T):
    a, b, c = map(int, input().split())
    print(shortest_referral_chain(a, b, c))
```

```python
import heapq
from collections import Counter

def count_unique_max_numbers(nums):
    freq = Counter(nums)
    max_heap = [-num for num in freq]
    heapq.heapify(max_heap)

    count = 0

    while max_heap:
        current = -heapq.heappop(max_heap)
        if freq[current] == 0:
            continue  # Skip already-removed numbers

        if freq[current] == 1:
            # Unique max
            count += 1
            freq[current] = 0
            half = current // 2
            if half > 0:
                freq[half] += 1
                heapq.heappush(max_heap, -half)
        else:
            # Repeated max, remove all
            freq[current] = 0

    return count

# Example usage:
n = int(input())
nums = list(map(int, input().split()))
print(count_unique_max_numbers(nums))
```

```python
def merge_intervals(intervals):
    intervals.sort()
    merged = []

    for start, end in intervals:
        if not merged or merged[-1][1] < start:
            merged.append([start, end])
        else:
            merged[-1][1] = max(merged[-1][1], end)
    
    return merged

def min_interval_difference(intervals):
    merged = merge_intervals(intervals)
    
    min_diff = float('inf')
    for i in range(1, len(merged)):
        diff = merged[i][0] - merged[i - 1][1]
        min_diff = min(min_diff, diff)
    
    return min_diff if len(merged) > 1 else 0

# Input reading
n = int(input())
intervals = [list(map(int, input().split())) for _ in range(n)]

print(min_interval_difference(intervals))
```
