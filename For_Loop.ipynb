***When ever you want to implement Counter program. initial count = 0 inside the function ***

```python
num = eval(input("Enter the number you want to check! : "))


def counter():
    d_count = 0               
    nd_count = 0
    for i in range(1, num+1):
        if num%i == 0:
            print("{} is divisible by {}".format(num,i))
            d_count = d_count +1

        else:
            print("{} is not divisible by {}".format(num,i))
            nd_count = nd_count + 1
    return(d_count,nd_count)
val1,val2 = counter()
print(val1,val2)
```

***FOR LOOP***

***for i in range(10):    
    ---- in the bracket if we have only single value then it is considered as stop value
    ---- python index always start from 0
    ---- so the default start value is=0
    ---- if direction sign is not mentioned then by default increment operation will proceed
    ---- if direction is positive side then end = stop-1***

***Pattern - 1***

```python
for i in range(10):
    print(i)
    #start = 0
    #direction +
    #stop = 10 - 1 = 9
#Output = 0 to 9    
    
```

$Pattern - 3 $

$for i in range(start, stop, step):
    ---- start : start of the loop, initial point
    ---- direction : what is sign of step value, that is direction
    ---- direction will not decide by start and stop value
    ---- direction will provided by step value
---- if step size is positive direction:
            end = stop - 1
---- if step size is negative direction:
            end = stop + 1$

```python
for i in range(2,20,2):
    print(i, end = ',')
    #start = 2
    #direction : step size = +2
    #end = stop - 1,  20-1 = 19
```

```python
for i in range(-2,-20,-2):
    print(i, end = ',')
    #start = 2
    #direction = step size = -2
    #end = stop+1, -20+1 = -19
```
