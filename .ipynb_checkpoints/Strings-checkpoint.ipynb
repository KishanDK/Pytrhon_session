```python
import string
dir(string)
```

```python
string.ascii_uppercase
```

```python
alpa = string.ascii_uppercase
print("The ascii_uppercase values are : ")
for i in alpa:
    print("The ASCII value of {} is {}".format(i,ord(i)))
```

```python
alp = string.ascii_lowercase
print("The ascii value for lowercase alphabets are : ")
for i in alp:
    print("The ascii value of {} is {}".format(i,ord(i)))
```

```python
for i in string.punctuation:
    print("The ascii values of {} is {}".format(i,ord(i)))
```

```python
# To find where the ascii values are start from
for i in range(1,50):
    print(i,chr(i))    # So 33 is the starting of ascii numbers
```

```python
# To find where the ascii value ends
for i in range(33,150):
    print(i,chr(i))        # 126 is the last ascii value
```

```python
#WAP ask the user find number of 'a' letter in a given string
# string = hai how are you and how do  you do ?

# 1st iterate the string using in operator and for loop 
# print each letter =====> i
# if condition : i == 'a' then count
string1 = "hai how are you and how do  you do ?"
count = 0
for i in string1:
    if i =='a':
        count = count+1
print(count)
```

```python
# WAP count how many vowels are their in the given string
# string2 = "hai how are you and how do  you do ?"

string2 = "hai how are you and how do  you do ?"
count = 0
for i in string2:
    #print(i)
    if i in 'aeiou':
        count = count+1
print(count)
```

```python
# WAP count how many unique vowels are their in the given string
# string2 = "hai how are you and how do  you do ?"
string2 = "hai how are you and how do  you do ?"
string3 = ''
for i in string2:
    #print(i)
    if i in 'aeiou' and i not in string3:
        string3 = string3+i
           
print(len(string3))
```
