```python
def add(a,b):
    print("Addition of {} and {} is {}".format(a,b,a+b))
def sub(a,b):
    print("Subraction of {} abd {} is {}".format(a,b,a-b))
def mul(a,b):
    print("Multiplication of  {} and {} is {}".format(a, b,a*b))
def div(a,b):
    print("Division of {} and {} is {}".format(a,b,a/b))

def main():
    print("If u want to enter 1 addition operation will come")
    print("If u want to enter 2 subraction operation will come")
    print("If u want to enter 3 multiplication operation will come")
    print("If u want to enter 4 division operation will come")
main()
operation = input("Enter any num [1,2,3,4] for operations")
a = eval(input("Enter a number"))
b = eval(input("Enter another number"))

if operation == '1':
    add(a,b)
elif operation == '2':
    sub(a,b)
elif operation == '3':
    mul(a,b)
elif operation == '4':
    div(a,b)
else:
    print("invalid operation!")
```
