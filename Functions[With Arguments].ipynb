```python
#Function with defauly arguments
import math
def area(r, pi = math.pi):     # default parameter
    area = round(pi * r * r , 2)
    print("The area of a circle is",area)
area(eval(input("Enter the radius of a circle : ")))
```

```python
# Return Concepts
import math

def area_of_circle():

    pii=math.pi
    
    r=eval(input("enter the radius"))
    
    area11=round(pii*r*r, 2)
    
    print("the area of circle is:", area11)
    
    return (area11, pii)
    
Area, Pi=area_of_circle()
    
print(Area)
    
print(Pi)
```

```python
def greater (n1, n2, n3):

    if n1>n2 and n1>n3:
    
        print('{} is greater'.format(n1))
    
        return(n1)
    
    elif n2>n3:
    
        print("{} is greater".format(n2))
    
        return(n2)
    
    else:
    
        print("{} is greater".format(n3))
    
        return(n3)
    
greater (20,200,10)

# It is returning only one values

# whenever condition satis
```
